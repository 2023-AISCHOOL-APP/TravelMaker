```python
from selenium import webdriver as wb   # 통제
from selenium.webdriver.common.keys import Keys   # 입력을 위한 컴퓨터용 키보드
from selenium.webdriver.common.by import By   # 선택자 선택의 기준
from bs4 import BeautifulSoup as bs   
import time
import pandas as pd
import numpy as np
```

### TourAPI4.0 사이트에서 상세설명 크롤링

```python
driver = wb.Chrome()
driver.get('https://api.visitkorea.or.kr/#/useInforArea')
```

```python
soup = bs(driver.page_source,'lxml')
```

```python
# 현재페이지에서 n개의 상세설명 가져오는 함수
# 해당 사이트 한페이지에 가져올 수 있는 최대 상세설명은 200개
# 위 사이트가 다음페이지 가져오기 하려면 관광지와 지역을 다시 선택해줘야 한다
# 그래서 한페이지(200개)씩 크롤링 하기 위해 아래와 같은 함수 작성
def overview(n):
    summary_list=[]
    title_list=[]
    for i in range(n):
        # 관광지 이름
        title = soup.select('p.app-title')[i].text
        title_list.append(title)
        # 관광지 상세설명
        summary = soup.select('div.bot>ul>li>div')[i].text
        summary_list.append(summary)
    return pd.DataFrame({'title':title_list, 'overview':summary_list})
```

```python
# 관광지 선택
btn= driver.find_elements(By.CLASS_NAME,'link')[1]
btn.click()
time.sleep(1)
```

```python
# 상세설명 가져오기
summary = soup.select_one('div.summary>div').text
summary
```
